After looking at the above output, we can say there are missing values for LotFrontage( Linear feet of street connected to property) has 259 missing values, electrical has 1 missing value. Whereas Alley, Bsmt (Basement), FireplacesQuality, Garage(Type, YrBlt, Finish, Quality and Conditition), MasVnrType, Pool, Fence, MiscFeature are are missing due to they do not exist in the data. So we replace them with better description. 

```python
## Converting missing values to more descriptive description
house_data['Alley'].replace(to_replace = np.nan, value = 'No alley access', inplace = True)
house_data['BsmtQual'].replace(to_replace = np.nan, value = 'No basement', inplace = True)
house_data['BsmtCond'].replace(to_replace = np.nan, value = 'No basement', inplace = True)
house_data['BsmtExposure'].replace(to_replace = np.nan, value = 'No basement', inplace = True)
house_data['BsmtFinType1'].replace(to_replace = np.nan, value = 'No basement', inplace = True)
house_data['BsmtFinType2'].replace(to_replace = np.nan, value = 'No basement', inplace = True)
house_data['MasVnrType'].replace(to_replace = np.nan, value = 'No Masonry veneer', inplace = True)
house_data['MasVnrArea'].replace(to_replace = np.nan, value = 0, inplace = True)
house_data['FireplaceQu'].replace(to_replace = np.nan, value = 'No Fireplace', inplace = True)
house_data['GarageType'].replace(to_replace = np.nan, value = 'No Garage', inplace = True)
house_data['GarageYrBlt'].replace(to_replace = np.nan, value = 0, inplace = True)
house_data['GarageFinish'].replace(to_replace = np.nan, value = 'No Garage', inplace = True)
house_data['GarageQual'].replace(to_replace = np.nan, value = 'No Garage', inplace = True)
house_data['GarageCond'].replace(to_replace = np.nan, value = 'No Garage', inplace = True)
house_data['PoolQC'].replace(to_replace = np.nan, value = 'No Pool', inplace = True)
house_data['Fence'].replace(to_replace = np.nan, value = 'No Fence', inplace = True)
house_data['MiscFeature'].replace(to_replace = np.nan, value = 'No Misc Feature', inplace = True)


## Removing missing values
house_data.dropna(inplace = True)

## Removing Id from the data set
house_data_2 = house_data.iloc[:,1:]
## convert all categorical variables to dummy variables
##house_data_2.dtypes
df_dummies = pd.get_dummies(house_data_2)
pd.set_option('display.max_rows', None)
df_dummies.head()
```

```python
# Get Correlation of SalePrice with other variables
plt.figure(figsize = (100,8))
df_dummies.corr()['SalePrice'].sort_values(ascending = False).plot(kind = 'bar')

# draw a line to see all strong correlation and negative 0.6
plt.plot([0,100],[0.6,0.6],'k-')
plt.plot([200,300],[-0.5,-0.5],'k-')
plt.show()
```

OverallQual, GrLivArea, GarageCars, GarageArea, TotalBsmtSF, 1stFlrSF seem to be positively correlated (>0.6) with SalePrice. While, Ultilites_AllPub(all public utilities), ExterQual_TA(Average/Typical) seem to be negatively correlated with SalePrice. 

Interestingly, ExterQual_TA, KitchenQual_TA,are negatived correlated, it has still poorer quality Fa, Po (fair, poor). However, they are not negatively correlated.

We will explore those patterns for the above correlations below before we dive into modeling and identifying important variables. 

## Feature Engineering

### MSSubClass: Nominal Categorical variables
#### Now they are in numbers, convert them using oneHotEncoder

```python
# First convert numbers back to descriptive, made the MSSubClass into an xlsx file and stored in data file
MSSubClass_names = pd.read_excel('data/MSSubClass.xlsx', header = None)
MSSubClass_names= MSSubClass_names.rename(columns = {0:'MSSubClass', 1:'MSSubClass_name'})
#MSSubClass_names
df = df.join(MSSubClass_names.set_index('MSSubClass'),on='MSSubClass')
print(df)
```

```python
## Now they are in description, create columns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')

# fit and transform the MSSubClass_name Column
df = np.array(ct.fit_transform(df))

# resassign column names
columns_names = ct.get_feature_names()
df = pd.DataFrame(df, columns = columns_names)
print(df)
```
